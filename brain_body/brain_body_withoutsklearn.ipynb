```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import sklearn.linear_model as ln
# using jupyter notebook theme
import jupyterthemes.jtplot as jt
jt.style(theme='gruvboxd')
```

```python
df = pd.read_fwf('brain_body.txt')
df
```

```python
_brain = df[['Brain']]
_body = np.array(df['Body'])
```

```python
# visualization 
plt.figure (figsize=[18,13])
plt.scatter(_brain,_body,s=140)
plt.xlabel('Brain')
plt.ylabel('Body')
plt.xlim(0,700)
plt.ylim(0,700)
```

```python
x_0 = np.array(df['Brain'])
den = x_0.dot(x_0) - x_0.mean() * x_0.sum()
slope = (x_0.dot(_body)-_body.mean()*x_0.sum())/den
intercept =(_body.mean()*x_0.dot(x_0)-x_0.mean()*x_0.dot(_body))/den
data1 = np.array([3500]) 
predict = slope*x_0+ intercept
print(slope)
print(intercept)
```

```python
# visualization 
plt.figure (figsize=[18,13])
plt.scatter(_brain,_body,s=140)
plt.plot(_brain,predict,color ='red',lineWidth =4)
plt.xlabel('Brain')
plt.ylabel('Body')
plt.xlim(0,700)
plt.ylim(0,700)
plt.show()
```
