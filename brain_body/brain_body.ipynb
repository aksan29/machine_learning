```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn.linear_model as ln
import os
# using jupyter notebook theme
import jupyterthemes.jtplot as jt
jt.style(theme='gruvboxd')
```

```python
df = pd.read_fwf('brain_body.txt')
df
```

```python
df.rename(columns={'Brain':'vol_brain'},inplace=True)
# visualization
plt.figure(figsize=[19,14])
plt.scatter(df['vol_brain'],df['Body'],s=100)
plt.xlim(0,700)
plt.ylim(0,700)
plt.xlabel('vol_brain',fontsize = 20)
plt.ylabel('Body',fontsize = 20)
```

```python
# fit data with linear regression algorithm
reg = ln.LinearRegression()
reg.fit(df[['vol_brain']],df['Body'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# make a prediction
new_predict = reg.predict(df[['vol_brain']])
residual = df['Body']-new_predict
plt.figure (figsize=(15,9))
plt.scatter(df['vol_brain'],residual , s=140)
plt.xlim(0,600)
plt.ylim(-200,200)
```

```python
# make plotting
logy = np.log(new_predict)
plt.figure(figsize=[15,10])
plt.scatter(df['vol_brain'],df['Body'],s=100)
plt.plot(df['vol_brain'],logy,color ='red',lineWidth =4)
plt.xlabel('vol_brain',fontsize = 20)
plt.ylabel('Body',fontsize = 20)
```
