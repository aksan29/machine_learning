```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn.linear_model as ln
import csv # import csv file
import jupyterthemes.jtplot as jt # I import this because I use jupyter theme for mu notebook
jt.style(theme='gruvboxd') # i used this to set my plotting theme based on the jupyter theme
```

```python
# importing data 
df = pd.read_csv('weight-height.csv')
df.shape
```

```python
# generating data information
print(df.info())
print(df.describe())
```

```python
# getting specific the detail data
# count the number og gender
print(df)
print(df['Gender'].value_counts())
data_up = (df[df['Height']>66])
print(data_up['Gender'].value_counts())
```

```python
# visualization the data 
df.plot(kind='scatter' ,x='Height',y='Weight',color ='blue')
```

```python
males = df[df['Gender']=='Male'] # define the number of man
females= df[df['Gender']=='Female'] # define the number of woman
# try to visualization the group of man and woman
fig , ax =plt.subplots()
males.plot(kind='scatter', x='Height' , y='Weight', color ='blue' , ax=ax)
females.plot(kind='scatter',x='Height',y='Weight',color ='red',ax=ax)
```

```python
#predict the weight based on height
x =np.array(df['Height']).reshape(10000,1)
y =np.array(df['Weight']).reshape(10000,1)
```

```python
reg = ln.LinearRegression()
reg.fit(x,y)
reg.coef_
x1=reg.predict(x)
```

```python
males = df[df['Gender']=='Male'] # define the number of man
females= df[df['Gender']=='Female'] # define the number of woman
# try to visualization the group of man and woman
fig , ax =plt.subplots()
males.plot(kind='scatter', x='Height' , y='Weight', color ='blue' , ax=ax)
females.plot(kind='scatter',x='Height',y='Weight',color ='red',ax=ax)
plt.plot(x,x1,color='yellow')
```

```python
males = df[df['Gender']=='Male'] # define the number of man
females= df[df['Gender']=='Female'] # define the number of woman

plt.figure (figsize=[18,13])
plt.scatter(males['Height'],males['Weight'],color = 'blue')
plt.scatter(females['Height'],females['Weight'],color ='red')
plt.plot(x,x1,lineWidth=4,color='yellow')
```
