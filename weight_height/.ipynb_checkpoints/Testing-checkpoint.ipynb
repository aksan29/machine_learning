## Step2 Initially and Calculate loss function

```python
# Calculate with scikit learn
reg= ln.LinearRegression()

y1 = np.array(df['Weight']).reshape(10000,1)
x1 = np.array(df['Height']).reshape(10000,1)
reg.fit(x1,y1)
reg.coef_
reg.intercept_

males =df[df['Gender']=='Male']
females =df[df['Gender']=='Female']
plt.figure(figsize=[15,9])
plt.scatter(males['Height'],males['Weight'],color ='blue',label='male')
plt.scatter(females['Height'],females['Weight'],color ='red',label='female')
plt.plot(df['Height'],reg.predict(x1),color ='green',lineWidth=4)
plt.legend()
plt.show

```

```python
print('''
from the scikit learn we get
''')
print('the slope is ',reg.coef_)
print('the intercept is',reg.intercept_)
```

```python
# calculate based on build mode
m = 0
c = 0
L= 0.0001

n= float(len(df['Weight']))
y = np.array(df['Weight'])
x = np.array(df['Height'])

for i in range(len(y)):
    y_predict = m*x+c
    d_c = -2/n * sum(y-y_predict)
    d_m = (-2/n) *sum(x*(y-y_predict))
    # update value of m and value of c
    m = m-(L*d_m)
    c= c-(L*d_c)
y_predict = m*x+c
plt.figure(figsize=[15,9])
plt.scatter(males['Height'],males['Weight'],color ='blue',label='male')
plt.scatter(females['Height'],females['Weight'],color ='red',label='female')
plt.plot(df['Height'],y_predict,color='green',lineWidth = 4)
plt.legend()
plt.show
```

```python
print('''
from the building model we can get 
''')
print('the slope is ',m)
print('the intercept is',c)
```

### model 3

```python
y3 = np.array(df['Weight'])
x3 = np.array(df['Height'])
den = x3.dot(x3) - x3.mean() * x3.sum()
slope3 = (x3.dot(y3)-y3.mean()*x3.sum())/den
intercept3 =(y3.mean()*x3.dot(x3)-x3.mean()*x3.dot(y3))/den
predict3 = slope3*x3+ intercept3
plt.figure(figsize=[15,9])
plt.scatter(males['Height'],males['Weight'],color ='blue',label='male')
plt.scatter(females['Height'],females['Weight'],color ='red',label='female')
plt.plot(df['Height'],predict3,color ='green',lineWidth=4)
plt.legend()
plt.show
```

```python
print('''
from the building model we can get 
''')
print('the slope is ',slope3)
print('the intercept is',intercept3)
```
